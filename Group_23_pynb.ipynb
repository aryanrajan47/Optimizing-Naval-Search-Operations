```python
# Block 1
import numpy as np
import math
import matplotlib.pyplot as plt
import copy
```

```python
#Block 2
def plot_heatmap(matrix, title):
    plt.imshow(matrix, cmap='binary', interpolation='nearest')
    plt.title(title)
    plt.colorbar()
    plt.show()
```

```python
#Block 3
n = int(input(f"No. of rows/columns of the input matrix: "))
```

```python
# Block 4
inp_array = np.zeros((n,n))

for i in range(n):
    for j in range(n):
        value = float(input(f"Enter value for position ({i}, {j}): "))
        inp_array[i][j] = value
inp_array
```

```python
#Block 5
# If you want to additionally change the input array values
array = copy.deepcopy(inp_array)
array[0][2] = 0.08
# array[6][3] = 0.2
# array[7][1] = 0.4
```

```python
#block 6
count = 0
rows = len(array)
columns = len(array)
for x in array:
  for y in x:
    if(y > 0):
      count = count +1
pos = np.zeros((3, count))
con = 0
for i in range(rows):
  for j in range(columns):
    if(array[i][j] > 0):
      pos[0][con] = i*10
      pos[1][con] = j*10
      pos[2][con] = array[i][j]
      con += 1

print('No.of values in mat:',count,'\n\npos:\n' , pos, '\n\nmat\n', array)
```

```python
#block 7
magnitude = int(input(f"Magnitude of the velocity: "))
theta = float(input(f"Angle of the velocity (randius): "))
```

```python
#Block 8
c = math.pi/180
vx = magnitude*np.cos(theta*c)
vy = magnitude*np.sin(theta*c)
print(vx,vy)
```
